```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def imshowww(name, img):
    b,g,r = cv2.split(img)
    img = cv2.merge([r,g,b])
    plt.title(name)
    plt.imshow(img)
    plt.show()

cv2.imshow = imshowww
```

```python
## 縮放圖片
img = cv2.imread(r"C:\Users\yifun\Desktop\tool_photo\sample.jpg") # cv2的檔案路徑無法接受任何中文字

print(type(img))
print(img.shape)
```
