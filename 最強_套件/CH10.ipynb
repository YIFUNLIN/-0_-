```python
import numpy as np
import matplotlib.pyplot as plt

days = np.arange(1, 11)	
weight = np.array([10, 14, 18, 20, 18, 16, 17, 18, 20, 17])

plt.ylim([0, weight.max()+1])	
plt.xlabel('days')
plt.ylabel('weight')

plt.plot(days, weight, marker='o', markerfacecolor='black')	

plt.show()
```

```python
x = [1,2,3,4,5]
y = [11,12,13,14,15]

labels = ['LA','BULLUS','Clipers','OKC','Warriors']
plt.bar(x, y, tick_label = labels)
plt.show()
```

```python
np.random.seed(0)
data = np.random.randn(10000)

plt.hist(data, bins='auto', density=True)
plt.show()
```

```python
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
c = np.random.choice(np.arange(100), 100)

plt.scatter(x, y, s=c, c=c, cmap='viridis')


plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(-2 * np.pi, 2 * np.pi)		
x, y = np.meshgrid(t, t)				
z = np.sin(np.sqrt(x ** 2 + y ** 2))		


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')		# 3D 

ax.plot_surface(x, y, z)   #

plt.tight_layout()
plt.show()
```
