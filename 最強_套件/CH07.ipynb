```python
import numpy as np
import pandas as pd

def make_random_df(index, columns, seed):
    np.random.seed(seed)
    df = pd.DataFrame()
    for c in columns:
        df[c] = np.random.choice(range(1,101),len(index))
    df.index = index
    return df

col = ['warriors','Lakers','BULLUS']
df1 = make_random_df(range(1,5),col,0)
df2 = make_random_df(range(1,5),col,1)

print(df1)

print(df2)
```

```python
df = pd.concat( [df1,df2],axis=0)
df
```

```python
df_0  = pd.concat([df1,df2],axis=1)
df_0
```

```python
dff = pd.concat([df1, df2], axis=1, keys=["X", "Y"])
print(dff)
```

```python
d1 = {"fruits": ["apple", "orange", "banana", "strawberry",
  "kiwifruit"],
  "year": [2001, 2002, 2001, 2008, 2006],
  "amount": [1, 4, 5, 6, 3]}
d1 = pd.DataFrame(d1)

d2 = {"fruits": ["apple", "orange", "banana", "strawberry",
  "mango"],
  "year": [2001, 2002, 2001, 2008, 2007],
  "price": [150, 120, 100, 250, 3000]}
d2 = pd.DataFrame(d2)

print(d1)
print(d2)
```

```python
df3 = pd.merge(d1,d2,on = 'fruits',how = 'inner')
df3
```
