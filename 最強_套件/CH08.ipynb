```python
import numpy as np

np.random.seed(0)

columns = ["apple", "orange", "banana", "strawberry","kiwifruit"]

df = pd.DataFrame()
for c in columns:
    df[c] = np.random.choice(range(1,11),10)

df.index = [i for i in range(1,11)]
df
```
