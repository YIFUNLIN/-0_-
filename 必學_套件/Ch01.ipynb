```
n = 9
print('檢查數字...')
if n >= 10:
  print('數字超過門檻')
else:
  print('數字未達門檻')
```

```
n = 16
if n >= 20:
  print(n, '是很大的數字')
elif n >= 15:     #n >= 15 但< 20時
  print(n, '是中等的數字')
elif n >= 10:     #n >= 10 但< 15時
  print(n, '是普通的數字')
else:           #n < 10 時
  print(n, '是較小的數字')
```

```
year = 2020

if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
  print(year, '是閏年')
else:
  print(year, '是平年')
```
